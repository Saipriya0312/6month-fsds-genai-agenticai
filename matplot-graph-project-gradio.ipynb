```python
import pandas as pd

data = {
    'Employee_Name'     : ['Saipriya','Sreehaan','Sreeshnav','Narsi','Vara','Kittu','Sindhu','Daksh'],
    'Employee_Sal'      :[20000,18000,19000,17500,16000,15000,15500,19000],
    'Employee_Expenses' :[12000,10400,10700,16700,14500,14300,13500,16200]
}

df = pd.DataFrame(data)
print(df)
```

```python
df[['Employee_Name','Employee_Sal']]
```

Line Graph

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
plt.plot(df['Employee_Name'],df['Employee_Sal'],color = 'blue',marker = 'o',ls = '-',label = 'Employee_Name')
plt.title("Employee Sals over Names")
plt.xlabel("Employee_Name")
plt.ylabel("Employee_Sal")
plt.grid('True')
plt.legend()
plt.show()
```

Bar Graph


```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
width =0.5
plt.bar(df['Employee_Name'],df['Employee_Sal'],width=width,color ='green',label = 'Employee_Sal')
plt.bar(df['Employee_Name'],df['Employee_Expenses'],width=width,color = 'black',label = 'Employee_Expenses',bottom=df['Employee_Sal'])
plt.title("Employee Expenses over Salaries")
plt.xlabel("Employee_Name")
plt.ylabel("Employee_Sal")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from enum import auto
plt.Figure(figsize=(6,5))
plt.pie(df['Employee_Expenses'],labels= df['Employee_Name'],autopct='%1.2f%%',startangle= 140,colors= plt.cm.Paired.colors)
plt.title('Expenses over Name')
plt.show()
```

Scatter Plot


```python
plt.Figure(figsize=(6,5))
plt.scatter(df['Employee_Name'],df['Employee_Expenses'],s =100,color = 'brown',edgecolors='black')
plt.title("Employee Expenses over Name")
plt.xlabel("Employee_Name")
plt.ylabel("Employee_Expenses")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Histogram


```python
plt.figure(figsize=(8,4))
plt.hist(df['Employee_Sal'],bins= 5,color = 'brown',edgecolor='black')
plt.title("Histogram of Employee Expenses over Salries")
plt.xlabel("Employee_Sal")
plt.ylabel("Employee_Expenses")
plt.tight_layout()
plt.show()
```

Box plot

```python
plt.figure(figsize=(8,4))
plt.boxplot(df['Employee_Sal'],vert= False,patch_artist= True,boxprops=dict(facecolor = 'lightgreen'))
plt.title("Salaries box plot")
plt.xlabel("Salaries")
plt.ylabel("Expenses")
plt.tight_layout()
plt.show()
```

```python
import gradio as gr
import pandas as pd
import matplotlib.pyplot as plt

def generate_plot(plot_type):
    fig = plt.figure(figsize=(8,5))
    
    if plot_type == 'Line Plot':
        plt.plot(df['Employee_Name'],df['Employee_Sal'],color = 'blue',marker = 'o',ls = '-',label = 'Employee_Name')
        plt.title("Employee Sals over Names")
        plt.xlabel("Employee_Name")
        plt.ylabel("Employee_Sal")
        plt.grid('True')
        plt.legend()
        plt.show()
    
    elif plot_type == 'Stacked Bar Plot':
        width =0.5
        plt.bar(df['Employee_Name'],df['Employee_Sal'],width=width,color ='green',label = 'Employee_Sal')
        plt.bar(df['Employee_Name'],df['Employee_Expenses'],width=width,color = 'black',label = 'Employee_Expenses',bottom=df['Employee_Sal'])
        plt.title("Employee Expenses over Salaries")
        plt.xlabel("Employee_Name")
        plt.ylabel("Employee_Sal")
        plt.legend()
        plt.grid(True)
        plt.show()
        
    elif plot_type =='Pie Plot':
        plt.pie(df['Employee_Expenses'],labels= df['Employee_Name'],autopct='%1.2f%%',startangle= 140,colors= plt.cm.Paired.colors)
        plt.title('Expenses over Name')
        plt.show()
        
    elif plot_type == 'Scatter Plot':
        plt.scatter(df['Employee_Name'],df['Employee_Expenses'],s =100,color = 'brown',edgecolors='black')
        plt.title("Employee Expenses over Name")
        plt.xlabel("Employee_Name")
        plt.ylabel("Employee_Expenses")
        plt.legend()
        plt.grid(True)
        plt.show()
        
    elif  plot_type == 'Histogram':
        plt.hist(df['Employee_Sal'],bins= 5,color = 'brown',edgecolor='black')
        plt.title("Histogram of Employee Expenses over Salries")
        plt.xlabel("Employee_Sal")
        plt.ylabel("Employee_Expenses")
        plt.show()

    elif  plot_type == 'Box Plot':
        plt.boxplot(df['Employee_Sal'],vert= False,patch_artist= True,boxprops=dict(facecolor = 'lightgreen'))
        plt.title("Salaries box plot")
        plt.xlabel("Salaries")
        plt.ylabel("Expenses")
        plt.show()
        
    plt.tight_layout()    
    return fig

#gradio UI
demo = gr.Interface(
    fn = generate_plot,
    inputs = gr.Radio(
            ['Line Plot','Stacked Bar Plot','Pie Plot','Scatter Plot','Histogram','Box Plot'],
            label = 'Choose Plot Type'),
    outputs = gr.Plot(label = 'Visualization'),
    title = 'Sales-Profit visual Explorer',
    description = 'Choose a plot type to visualize the data'
)

demo.launch()
```
