```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
```

Scatter plot

```python
sns.scatterplot(data=tips,x = 'total_bill',y = 'tip',hue='time',size='size',palette='deep')
```

Line Plot

```python
sns.lineplot(data=tips,x = 'size',y = 'total_bill',hue='sex',ci=None,marker='o')
plt.title('line plot of toatl bill against size')
```

Bar plot

```python
sns.barplot(data=tips,x = 'day',y = 'total_bill',hue='sex',palette='muted')
plt.title('bar plot of day vs total bill')
```

Box Plot

```python
sns.boxplot(data=tips,x= 'day',y ='tip',hue = 'smoker',palette='dark')
plt.title("")
```

```python
tips.columns
```

countPlot

```python
sns.countplot(data= tips,x = 'day',hue='smoker',palette= 'dark')
plt.title('count plot of days by smokers')
```

Regression plot

```python
sns.regplot(data= tips,x = 'total_bill',y = 'tip',scatter_kws={'s':50},line_kws={'color':'Red'})
plt.title("Regression plot of total bill vs tip")
plt.show()
```

Histogram

```python
sns.histplot(data = tips,x = 'total_bill',bins= 20,kde=True,color= 'blue')
plt.title('Histogram plot of total bill with KDE')
plt.show()  #positive scew
```

Paired plot

```python
sns.pairplot(tips,hue= 'sex',vars = ['total_bill','tip','size'],palette= 'husl')
plt.suptitle('Numerical varibles by Gender',y = 1.02)
plt.show()
```

CatPlot

```python
tips.columns
```

```python
sns.catplot(data=tips, x = 'day',y = 'tip',hue= 'sex',kind='point',palette= 'bright')
plt.title('CatPlot of tips by day and gender')
```

Joint Plot

```python
sns.jointplot(data = tips, x = 'total_bill',y = 'tip',hue = 'sex',palette='coolwarm',kind='scatter')
plt.suptitle("JointPlot of total bill by sex and tip")
```

FacetGrid

```python
g = sns.FacetGrid(tips,col = 'time',row= 'smoker',margin_titles=True).map(sns.histplot,'total_bill',bins = 20,kde=True,color= 'blue').add_legend()
g
```

Strip Plot

```python
sns.stripplot(data= tips,x = 'day',y = 'tip',hue= 'sex',jitter= True,palette= 'Set1')
plt.title('Strip plot of tip based on gender and day')
plt.show()
```

KDE plot

```python
sns.kdeplot(data = tips,x = 'total_bill',hue = 'sex',fill = True,palette= 'tab10')
plt.title('KDE plot of total bill by gender')
plt.show()
```

Violin plot

```python
sns.violinplot(data= tips,y ='total_bill',x = 'day',hue= 'time',split= True,palette='pastel')
plt.title('Violin plot of total bill base don gender and tip')
plt.show()
```
